```python
def unique_values(df):
  for column in df:
    if df[column].dtypes =='object':
      print(f'{column} : {df[column].unique()}')
```

```python
yes_no_cols = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
```

```python
for col in yes_no_cols:
  df1[col].replace({'Yes':1,'No':0},inplace=True)
```

```python
unique_values(df1)
```

```python
df1['gender'].replace({'Female':0,'Male':1},inplace=True)
```

```python
df2=pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod'])
```

```python
df2.columns
```

```python
df2.dtypes
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
df2
#we are brining the selected columns to a scale of 0 to 1
```

```python
#now we start train_test_split
```

```python
x=df2.drop('Churn',axis='columns')
y=df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
len(x_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),   #we are creating a nueral network model. Here we create an input and output layer,hidden layer isnt ccompulsory
    keras.layers.Dense(1,activation='sigmoid'),
])
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train,y_train)
```

```python
model.fit(x_train,y_train,epochs=5)
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
yp=model.predict(x_test)
yp
```

```python
y_pred=[]
for element in yp:
  if element>0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred)) #below will give overall performance of the model
```

```python
import seaborn as sn
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
```

```python
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
